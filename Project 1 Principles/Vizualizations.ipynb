# _ _ _ _**Importing Files**_ _ _ _

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the data from the Google Sheets link
url = "/content/Sales Pred [Week(2)].csv"
df = pd.read_csv(url)

# Printing the first 5 rows of the dataframe
print(df.head())
```

#_ _ _ _**Droping Duplicates**_ _ _ _

```python
# Counting duplicates
print(f"We have {df.duplicated().sum()}")

# Droping duplicate rows
df.drop_duplicates(inplace=True)

# Verify the number of rows after dropping duplicates
print("Number duplicates after:", df.duplicated().sum())
```

#_ _ _ _**Cleaning NaNs**_ _ _ _

```python
# Replace NaNs with the mode of the corresponding column
df.fillna(df.mode().iloc[0], inplace=True)

# Verify that there are no more NaNs in the dataframe
print("Number of NaNs after cleaning:", df.isna().sum().sum()) # BY USING sum() TWO TIMES I GET THE TOTAL COUNTING ALL THE ROWS
```

# _ _ _ _**Visualizations**_ _ _ _

#_ **Bar Chart: Item Type Distribution**

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Item_Type")
plt.xticks(rotation=90)
plt.grid()
plt.title("Distribution of Item Types")
plt.show()
```

**This will create a bar chart showing the distribution of different item types in the dataset.**
- **Note:** By "distribution" in this context, I mean the frequency or count of each item type in the dataset

# **Box Plot: Item MRP by Outlet Type**

```python
plt.figure(figsize=(10,12))
sns.boxplot(data=df, x="Outlet_Type", y="Item_MRP")
plt.grid()
plt.title("Item MRP by Outlet Type")
plt.show()
```

This will create a box plot showing the distribution of item MRP (maximum retail price) by different outlet types.

_ **Pair Plot:**

```python
sns.set(style="ticks")
sns.pairplot(data=df, vars=["Item_Weight", "Item_Visibility", "Item_MRP"])
plt.show()
```

This will create a grid of plots with scatterplots of each pair of variables **(Item_Weight, Item_Visibility, and Item_MRP)** on the diagonal, and histograms of each variable on the other axes. This type of plot is useful for exploring the relationships between multiple variables in a dataset.
